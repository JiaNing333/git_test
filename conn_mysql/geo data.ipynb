```python
import pandas as pd
import numpy as np
```

```python
cafe_point = pd.read_csv('C:/Users/Student/Desktop/Project/data/chain_cafe.csv')
```

```python
cafe_point.tail()
```

```python
points = [{'type': 'Feature',
  'properties': {'type':'chain_cafe', 'brand':'cama', '_id': 0},
  'geometry': {'type': 'Point',
   'coordinates': [121.52973175048827, 25.029772632953623]}},
 ]
```

```python
points[0]['geometry']['coordinates'][0] = cafe_point.loc[0,'long']
points[0]['geometry']['coordinates'][1] = cafe_point.loc[0,'lat']
```

```python
points
```

```python
# 加入咖啡店資料
for i in range(1, cafe_point.shape[0]): #cafe_point.shape[0]
    tmp = {'type': 'Feature',
           'properties': {'type':'NA', 'brand':'NA', '_id': 0},
           'geometry': {'type': 'Point','coordinates': [0, 0]}}
    tmp['properties']['type'] = 'chain_cafe'
    tmp['properties']['brand'] = cafe_point.loc[i,'brand']
    tmp['properties']['_id'] = i
    tmp['geometry']['coordinates'][0] = cafe_point.loc[i,'long']
    tmp['geometry']['coordinates'][1] = cafe_point.loc[i,'lat']
    points.append(tmp)
```

```python
rent_point = pd.read_csv('C:/Users/Student/Desktop/Project/data/rental.csv')
```

```python
rent_point.tail()
```

```python
# 加入出租點資料
for i in range(rent_point.shape[0]): #rent_point.shape[0]
    tmp = {'type': 'Feature',
           'properties': {'type':'NA', 'brand':'NA', '_id': 0},
           'geometry': {'type': 'Point','coordinates': [0, 0]}}
    tmp['properties']['type'] = 'rent'
    tmp['properties']['brand'] = ''
    tmp['properties']['_id'] = i
    tmp['geometry']['coordinates'][0] = rent_point.loc[i,'long']
    tmp['geometry']['coordinates'][1] = rent_point.loc[i,'lat']
    points.append(tmp)
```
