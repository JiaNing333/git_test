```python
import pandas as pd
import numpy as np
```

```python
conn = pymysql.connect(**db_settings)

with conn.cursor() as cursor:
    
    command = pd.read_sql("select `brand`, `addr`, `lat` , `long` ,`2019_income` from cafe_Feature_500m_CSV;",con=conn).decode()
    
    cursor.execute(command)
    
    result = cursor.fetchall()
    
    i_result = result.__iter__()
    
    for i in i_result:
            print(list(i))
```
