```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
url = 'https://www.findcoupon.tw/showroom-1076/store/%E5%8F%B0%E5%8C%97%E5%B8%82'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

# print(soup)

```

```python
st = soup.find_all("div",class_ = 'tr')
print(st)
print(type(st))
```

```python
import json
import pprint as pp
stlist = []

st = soup.select("div.tr > div.td")
# print(soup.select_one("div.tr > div.td").get_text())

for sts in st:
    stlist.append(sts.get_text())
# print(stlist)

def split_list(stlist, n):
    for idx in range(0,len(stlist),n):
        yield stlist[idx:idx+n]
        
stbks = list(split_list(stlist,3))
pp.pprint(stbks)

def saveJson():
    file = open("starbacks.json", "w", encoding="utf-8")
    file.write(json.dumps(stbks, ensure_ascii=False))
    file.close() 
    
if __name__ == "__main__":
    saveJson()
```
