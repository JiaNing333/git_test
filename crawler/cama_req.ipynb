```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.fonfood.com/store/542399/branch/%E5%8F%B0%E5%8C%97'

response = requests.get(url)
# print(response.text)
# soup = BeautifulSoup(page_source, 'html.parser')
soup = BeautifulSoup(response.text, 'html.parser')
# print(soup)
```

```python
cama = soup.select('a.darkOrange')
print( soup.select_one('a.darkOrange').get_text())
```

```python
import json
import pprint as pp
camalist = []
all_ = soup.select('#store > div.bodyPanel > div.sectionBody > table > tbody> tr> td')

for alls in all_ :   
    camalist.append(alls.get_text().replace("\n",""))
# print(camalist)

def split_list(camalist, n):
    for idx in range(0,len(camalist),n):
        yield camalist[idx:idx+n]

camas = list(split_list(camalist,4))

pp.pprint(camas)

def saveJson():
    file = open("cama.json", "w", encoding="utf-8")
    file.write(json.dumps(camas, ensure_ascii=False))
    file.close() 
    
if __name__ == "__main__":
    saveJson()
```
