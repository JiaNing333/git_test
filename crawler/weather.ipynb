```python
import requests,json
```

```python
# CWB-62EE07E1-8959-4BC8-9820-B532C37AF8BC
# https://opendata.cwb.gov.tw/fileapi/v1/opendataapi/F-D0047-091?Authorization=CWB-62EE07E1-8959-4BC8-9820-B532C37AF8BC&downloadType=WEB&format=JSON

auth_code = 'CWB-62EE07E1-8959-4BC8-9820-B532C37AF8BC'

r=requests.get(f"https://opendata.cwb.gov.tw/fileapi/v1/opendataapi/F-D0047-091?Authorization={auth_code}&downloadType=WEB&format=JSON")

# print(r.text) 資料很大,記憶體位置如果不夠大,要注意可能會當機

obj = json.loads(r.text)
# ['datasetInfo'] 裡面的資料不是我要的
for objLocation in obj ['cwbopendata']['dataset']['locations']['location'] :
    print(f"縣市名稱: {objLocation['locationName']}, lat: {objLocation['lat']}, lon: {objLocation['lon']}")

print('='*50)

for objLocation in obj ['cwbopendata']['dataset']['locations']['location'] :
    print(f"-*-*-*縣市名稱:{objLocation['locationName']}-*-*-*")
    for objWeatherElement in objLocation ['weatherElement']:
        if objWeatherElement['elementName'] == 'T' :
            print(f"description: {objWeatherElement['description']}")
            for objtime in objWeatherElement['time']:
                print(f"startTime: {objtime['startTime']}, endTime: {objtime['endTime']}")
                print(f"溫度: {objtime['elementValue']['value']} {objtime['elementValue']['measures']}")


```
